It can be seen that there is multicollinearity between several features such as radius, area, and perimeter. In addition, there are also features such as compactness, concavity, and concave point which also implies multicollinearity.

```python
#look at the correlation matrix of them with heatmap (bcs they're multivar)

# Generate and visualize the correlation matrix
corr = data.corr().round(2)

# Mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set figure size
f, ax = plt.subplots(figsize=(20, 20))

# Define custom colormap
cmap = sns.diverging_palette(240, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.tight_layout()
```

```python
# Draw the heatmap again, with the new correlation matrix
corr = data.corr().round(2)
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corr, mask=mask, cmap=cmap, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
plt.tight_layout()
```

<h3> Split Data </h3>

```python
X = data.drop(['diagnosis'],axis=1)
Y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=5)
```

<h3> Synthetic Minority Oversampling Technique (SMOTE) </h3>

```python
from imblearn.over_sampling import SMOTE

#memisah nama kolom variabel prediktor dan target
y_cols=["diagnosis"]
x_cols= [x for x in data.columns if x not in y_cols]

#handle imbalance class using oversampling minority class with smote method
os = SMOTE(sampling_strategy='minority',random_state = 123)
train_smote_X,train_smote_Y = os.fit_resample(X_train,Y_train)
train_smote_X = pd.DataFrame(data = train_smote_X,columns=x_cols)
train_smote_Y = pd.DataFrame(data = train_smote_Y,columns=y_cols)
```

```python
#proportion before using SMOTE
Y_train.value_counts()
```

```python
#proportion after using SMOTE
train_smote_Y["diagnosis"].value_counts()
```

<h3> Feature Scalling </h3>

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()

X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

<h3> Model Data With Random Forest </h3>

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

rfc= RandomForestClassifier()
model = rfc.fit(X_train, Y_train)
prediction = model.predict(X_test)
confusion_matrix(Y_test, prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, prediction)
```

```python
from sklearn.metrics import classification_report
```

```python
y_train_pred = rfc.predict(X_train)
y_test_pred = rfc.predict(X_test)

print('RANDOM FOREST TRAINING RESULT:')
print('======================================================')
print('Accuracy:', rfc.score(X_train, Y_train))
print('======================================================')
print('Classification Report --------------------------------')
print(classification_report(Y_train, y_train_pred))
print('======================================================')
print('Confussion Matrix ------------------------------------')
print(confusion_matrix(Y_train, y_train_pred))

print('\n======================================================')

print('\n RANDOM FOREST TESTING RESULT:')
print('======================================================')
print('Accuracy:', rfc.score(X_test, Y_test))
print('======================================================')
print('Classification Report --------------------------------')
print(classification_report(Y_test, y_test_pred))
print('======================================================')
print('Confussion Matrix ------------------------------------')
print(confusion_matrix(Y_test, y_test_pred))
```

<h3> Check Confusion Matrix </h3>

```python
y_pred = rfc.predict(X_test)
cf_matrix = confusion_matrix(Y_test, y_pred)

group_names = ["True Negative", "False Positive", "False Negative", "True Positive"]
group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]

labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)

plt.figure(figsize=(10, 8))
sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Pastel2_r')
```
